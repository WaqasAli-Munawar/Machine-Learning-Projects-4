Complex systems, such as a power grid, the World Wide Web, activity in different regions of the brain, or people within a community, can be understood, studied and visualized based on their connections in a network. 

Network Graph Analysis has real broad applications in the field of networking. Two main areas are involved in the analysis of the application of network graphs, which are a graph-based representation and network theory.

A graph has two components which are **nodes and edges**. In a graphical representation, these components have natural correspondences with the elements of the problem. In general, 
* nodes in a graph represent features and 
* edges represent interactions between features.

Most complex systems are problems of a graphical type, such as 
* the transportation network, 
* the trade links of an industry within a country, 
* the genetic interaction network, and 
* the Internet. 

On the other hand, network theory works as a tool that provides a set of techniques to analyze a graph and apply network theory using a graphical representation.

### Network Graph Analysis

For this network graph analysis, we will be using data from the tags used by Stack Overflow. The dataset we’re using here contains network links, source and target technical tags, and the link value between each pair.

It also contains the nodes of the network, the name of each node, the group to which this node belongs and a node size based on the frequency of use of this technological beacon. So let’s see how to visualize and analyze a network graph:

```python
import networkx as nx
import matplotlib.pyplot as plt
import pandas as pd

# Input data files check
from subprocess import check_output

import warnings
warnings.filterwarnings('ignore')
```

```python
G = nx.Graph(day="Stackoverflow")
df_nodes = pd.read_csv('stack_network_nodes.csv')
df_edges = pd.read_csv('stack_network_links.csv')
```

```python
for index, row in df_nodes.iterrows():
    G.add_node(row['name'], group=row['group'], nodesize=row['nodesize'])
    
for index, row in df_edges.iterrows():
    G.add_weighted_edges_from([(row['source'], row['target'], row['value'])])
```

```python
color_map = {1:'#f09494', 2:'#eebcbc', 3:'#72bbd0', 4:'#91f0a1', 5:'#629fff', 6:'#bcc2f2',  
             7:'#eebcbc', 8:'#f1f0c0', 9:'#d2ffe7', 10:'#caf3a6', 11:'#ffdf55', 12:'#ef77aa', 
             13:'#d6dcff', 14:'#d2f5f0'} 

plt.figure(figsize=(25,25))
options = {
    'edge_color': '#FFDEA2',
    'width': 1,
    'with_labels': True,
    'font_weight': 'regular',
}

colors = [color_map[G.nodes[node]['group']] for node in G]
sizes = [G.nodes[node]['nodesize']*10 for node in G]

"""
Using the spring layout : 
- k controls the distance between the nodes and varies between 0 and 1
- iterations is the number of times simulated annealing is run
default k=0.1 and iterations=50
"""
nx.draw(G, node_color=colors, node_size=sizes, pos=nx.spring_layout(G, k=0.25, iterations=50), **options)
ax = plt.gca()
ax.collections[0].set_edgecolor("#555555") 
plt.show()
```
